```python
import json
import pandas as pd
retrieval = ['FinDER', 'FinQABench', 'FinanceBench', 'TATQA', 'FinQA', 'ConvFinQA', 'MultiHiertt']
file_type = ['corpus', 'queries']
for t in file_type:
    for file in retrieval:
        jsonl_path_corpus = f"/Users/nattapat/ICAIF-FinanceRAG/dataset/{file}/corpus.jsonl"
        data = []
        with open(jsonl_path_corpus, "r") as f:
            for line in f:
                data.append(json.loads(line.strip()))
        df = pd.DataFrame(data)
        df.to_csv(f"dataset_csv/{file}_{t}.csv", index=False)
    
```
